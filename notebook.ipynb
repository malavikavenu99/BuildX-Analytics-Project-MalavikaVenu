```python
# Import pandas library
import pandas as pd
# Load the dataset
df = pd.read_csv("student.csv")
# Display the first five rows
df.head()
```

```python
# Check dataset information
df.info()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Missing salary values replaced with 0 because students who were not placed do not have salary records.
df["salary"] = df["salary"].fillna(0)
```

```python
# Checked and removed duplicate records to improve data quality.
df.duplicated().sum()
```

```python
# Clean text columns
text_columns = ["gender","ssc_b","hsc_b","hsc_s","degree_t","workex","specialisation","status"]
for col in text_columns:
    df[col] = df[col].str.strip().str.title()
```

```python
df.dtypes
```

```python
df["salary"] = df["salary"].astype(int)
```

```python
# Rename columns for consistency
df.rename(columns={
    "ssc_p":"ssc_percentage",
    "hsc_p":"hsc_percentage",
    "degree_p":"degree_percentage",
    "mba_p":"mba_percentage",
    "etest_p":"employability_test_percentage"
}, inplace=True)
```

```python
# Create performance category column
df["performance_category"] = df["degree_percentage"].apply(
    lambda x: "Excellent" if x >= 75 else
              "Good" if x >= 60 else
              "Average"
)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
